#### 7. Rename the column names appropriately

```python
# column names
print(df.columns)
```

```python
#renaming column names
df = df.rename(columns={'Mat': 'Matches', 
                        'Inns':'Innings', 
                        'Wkts':'Wickets',
                        'BBI':'Best_Innings_Bowling', 
                        'BBM': 'Best_Match_Bowling',
                        'Ave': 'Average', 
                        'Econ': 'Economy_rate', 
                        'SR':'Strike_rate', 
                        '5':'5_wkts_in_inns', 
                        '10':'10_wkts_in_inns'})
df.head()
```
